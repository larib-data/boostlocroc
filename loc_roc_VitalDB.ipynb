```python
import mne
import numpy as np
from boost_loc_roc.LoC_RoC import extract_loc_roc, Truncate_fif, plot_spectrogram
import pandas as pd
```

```python
# Placeholder function for converting EEG data to a format compatible with Truncate_fif and extract_loc_roc
def create_mne_raw(eeg_data_1, eeg_data_2, sampling_frequency):
    info = mne.create_info(
        ch_names=["EEG1", "EEG2"], sfreq=sampling_frequency, ch_types=["eeg", "eeg"]
    )
    raw = mne.io.RawArray([eeg_data_1, eeg_data_2], info)
    return raw


# Load the Parquet file
file_path = "/home/emiliaturc/Documents/Data_VitalDB/vitalDB_10.parquet"
df = pd.read_parquet(file_path)

# Sampling frequency in Hz
# sampling_frequency = 128

# Initialize an empty list to store the results
results = []

# Extract unique patient indices
unique_idx_values = df["idx"].unique()

# Assuming you have an MNE Raw object named 'raw'
desired_frequency = 63  # The desired frequency in Hz

# Process each patient's EEG data
for idx_value in unique_idx_values:
    print(f"Processing Patient {idx_value}")
    sub_df = df[df["idx"] == idx_value]
    eeg_data_channel_1 = sub_df["EEG1"].values
    eeg_data_channel_2 = sub_df["EEG2"].values
    datetime_array = sub_df["datetime"].values

    # Convert datetime array to seconds
    datetime_index = pd.to_datetime(datetime_array)
    time_in_seconds = (datetime_index - datetime_index[0]).total_seconds()
    sampling_frequency = len(time_in_seconds) / time_in_seconds[-1]

    # Create MNE Raw object
    raw = create_mne_raw(
        eeg_data_channel_1 / 10**6, eeg_data_channel_2 / 10**6, sampling_frequency
    )

    # Truncate the EEG data
    raw = Truncate_fif(raw)
    duration_sec = raw.n_times / raw.info["sfreq"]

    # Downsample the Raw object to the desired frequency
    # raw = raw.copy().resample(sfreq=desired_frequency, npad="auto")

    if duration_sec > 5 * 60:
        Fs = raw.info["sfreq"]  # Hz
        time_loc, time_roc, t_proba, proba = extract_loc_roc(raw)
        # Visualize Spectrogram, EEG, and proba
        plot_spectrogram(
            time_loc,
            time_roc,
            eeg_data_channel_2,
            sampling_frequency,
            time_in_seconds,
            t_proba,
            proba,
        )

        # Append the results
        results.append({"idx": idx_value, "loc_time": time_loc, "roc_time": time_roc})

# Convert results to a DataFrame
results_df = pd.DataFrame(results)
```

```python
# Save the results to a CSV file
results_df.to_csv("loc_roc_results_VitalDB.csv", index=False)

print("LOC and ROC times computed and saved to loc_roc_results.csv")
```
